## 1. Introducción a POO

La programación orientada a objetos organiza el código alrededor de **objetos**.

### Conceptos clave
- **Clase**: la plantilla.
- **Objeto**: una instancia concreta de esa plantilla.
- **Atributo**: la información del objeto.
- **Método**: lo que el objeto sabe hacer.

Piensa así:
- la clase describe,
- el objeto existe,
- los atributos guardan estado,
- los métodos expresan comportamiento.


```python
# Example 1: first class and first object
class Perro:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def ladrar(self, sonido="Guau"):
        print(f"{self.nombre} dice: {sonido}")


mi_perro = Perro("Fido", 3)
print(f"El perro se llama {mi_perro.nombre} y tiene {mi_perro.edad} años.")
mi_perro.ladrar()
```

## 2. Atributos, métodos y `__init__`

El método `__init__` se ejecuta automáticamente cuando creamos un objeto.

Sirve para inicializar sus atributos.

Los métodos usan `self` para acceder a los datos del propio objeto.


```python
import math


class Circulo:
    def __init__(self, radio):
        self.radio = float(radio)

    def area(self):
        return math.pi * (self.radio ** 2)


mi_circulo = Circulo(6)
print("Radio:", mi_circulo.radio)
print("Área:", mi_circulo.area())
```

## 3. Herencia

La herencia permite reutilizar código y crear jerarquías.

Una clase hija puede:
- heredar atributos y métodos,
- añadir atributos nuevos,
- sobrescribir métodos.

`super()` permite aprovechar el constructor o comportamiento de la clase padre.


```python
class Animal:
    def __init__(self, nombre, patas=4):
        self.nombre = nombre
        self.patas = patas

    def hablar(self):
        print(f"{self.nombre} hace un sonido.")


class Perro(Animal):
    def __init__(self, nombre, raza):
        super().__init__(nombre)
        self.raza = raza

    def hablar(self):
        print(f"{self.nombre} dice: ¡Guau!")


class Gato(Animal):
    def __init__(self, nombre, color):
        super().__init__(nombre)
        self.color = color

    def hablar(self):
        print(f"{self.nombre} dice: Miau")


perro = Perro("Rex", "Labrador")
gato = Gato("Michi", "Blanco")
perro.hablar()
gato.hablar()
```

```python
# Exercise 3

class Vehiculo:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

    def descripcion(self):
        print(f"Este es un {self.marca} {self.modelo}.")    

class Coche(Vehiculo):
    def __init__(self, marca, modelo, puertas):
        super().__init__(marca, modelo)
        self.puertas = puertas

    def descripcion(self):
        print(f"Este es un {self.marca} {self.modelo} con {self.puertas} puertas.")

coche = Coche("Toyota", "Corolla", 4)
coche.descripcion() 
```

## 4. Métodos especiales: `__str__`

Los métodos especiales permiten integrar mejor nuestros objetos con Python.

Hoy nos quedamos con uno muy útil:

- `__str__`: define cómo se ve el objeto cuando usamos `print()`.

Hace que nuestros objetos sean más legibles y más fáciles de depurar.


```python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def __str__(self):
        return f"{self.nombre}, {self.edad} años"

    def descripcion(self):
        return f"Persona: {self.nombre}, Edad: {self.edad}"


persona = Persona("Lucía", 28)
print("La persona es ",persona)
print(persona.descripcion())
```

## 5. Ejercicio final: sistema de gestión de estudiantes

Ahora vamos a combinar todo.

Crea las siguientes clases:

### `Persona`

### `Estudiante` (hereda de `Persona`)
- atributo extra: `carrera`

- `__str__` personalizado

### `Profesor` (hereda de `Persona`)
- atributo extra: `departamento`
- atributo extra: `asignaturas` (lista)
- método: `agregar_asignatura(nombre)`
- `__str__` personalizado

### Qué hacer después
- crea al menos dos estudiantes con notas
- crea un profesor
- añade asignaturas al profesor
- imprime toda la información


```python
# Final exercise

class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def __str__(self):
        return f"{self.nombre}, {self.edad} años"

class Estudiante(Persona):
    def __init__(self, nombre, edad, carrera, materias=None):
        super().__init__(nombre, edad)
        self.carrera = carrera
        self.materias = materias if materias is not None else []
    
    def __str__(self):
        return f"{self.nombre}, {self.edad} años, estudiante de {self.carrera}, materias {self.materias}"

    def promedio(self):
        if not self.materias:
            return 0
        return sum(self.materias) / len(self.materias)

class Profesor(Persona):
    def __init__(self, nombre, edad, departamento,asignaturas=None):
        super().__init__(nombre, edad)
        self.departamento = departamento
        self.asignaturas = asignaturas if asignaturas is not None else []
    
    def __str__(self):
        return f"{self.nombre}, {self.edad} años, profesor del departamento de {self.departamento}, asignaturas {self.asignaturas}"
    
    def agregar_asignatura(self, asignatura):
        self.asignaturas.append(asignatura)

estudiante1 = Estudiante("Ana", 22, "Ingeniería Informática", [8.5, 9.0, 7.5])
estudiante2 = Estudiante("Luis", 24, "Matemáticas",[7.0, 6.5, 8.0])

profesor = Profesor("Dr. García", 45, "Informática")
profesor.agregar_asignatura("Programación")
profesor.agregar_asignatura("Estructuras de Datos")

print(estudiante1)
print(estudiante2)
print(profesor)
```

```python
# Optional challenge

class Tienda:

    def __init__(self, nombre, direccion, num_empleados, ventas=None):
        self.nombre = nombre
        self.direccion = direccion
        self.num_empleados = num_empleados
        self.ventas = ventas if ventas is not None else []
    
   
    def calcular_ventas_totales(self):
        #Version usando bucles
        """       
        total = 0
        for venta in self.ventas:
            total += venta
        return sum(total)
        """
        #Version usando funcion sum
        return sum(self.ventas)
        
    def ventas_ultimo_mes(self):
        return self.ventas[-1]
    
    def media_mensual_por_empleado(self):
        if not self.ventas or self.num_empleados == 0:
            return 0
        # Media mensual = es la suma de todas las ventas  / el total de meses
        # Media mensual por empleado Mensual mensual / el total de empleados
        media_mesnual = self.calcular_ventas_totales() / len(self.ventas)
        return media_mesnual / self.num_empleados
    
tienda = Tienda("MarketPlace", "calle torrejon 26", 5, [42, 12, 45])
print(f"Venas totales: {tienda.calcular_ventas_totales()}")
print(f"Media Mensual por Empleados: {tienda.media_mensual_por_empleado()}")

```

```python
"""
Decorador @staticmethod :
- Se usa para definir un método dentro de una clase que no necesita acceder 
  ni a la instancia (self) ni a la clase (cls).
- Es una función “normal”, pero vive dentro de la clase por organización.
- No necesitas crear un objeto para llamarla Tienda.es_buen_mes(1500)  # → True
"""
class Tienda:

    @staticmethod
    def es_buen_mes(ventas):
        return ventas > 1000
    
```

```python
"""
Decorador @classmethod: 
Es un método que recibe la clase (cls) en lugar de la instancia (self).
    - Recibe cls (la clase) como primer parámetro
    - Puede acceder y modificar la clase
    - No necesita crear un objeto para usarse
    - Se puede usar como constructor alternativo
    - No creas el objeto directamente, lo hace el método usando cls.
Donde:
    self → es el objeto (instancia)
    cls → es la clase

Sirve para crear formas alternativas de construir objetos.
    Persona("Juan")              # nombre normal
    Persona.crear_sin_nombre()   # nombre automático o por defecto
"""

class Persona:

    def __init__(self, nombre):
        self.nombre = nombre

    @classmethod
    def crear_sin_nombre(cls):
        return cls("Anónimo")
p1 = Persona("Juan")
p2 = Persona.crear_sin_nombre()
print(p1.nombre)
print(p2.nombre)  # Anónimo 
```
