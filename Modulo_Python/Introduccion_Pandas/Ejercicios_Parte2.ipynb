```python
import numpy as np
import pandas as pd
from IPython.display import display

np.random.seed(42)
pd.set_option('display.max_columns', 60)
pd.set_option('display.width', 180)
```

Trabajaremos con tres tablas con problemas realistas:
- `customers`: dimensión de clientes
- `products`: dimensión de productos
- `transactions`: tabla de transacciones

La idea es limpiarlas y convertirlas en una tabla analítica única.

### DataFrame Customers
- Total de clientes: 120
- Columnas [customer_id, name, city, segment, age]

```python
# Crear DataFrame Customers
n_customers = 120
customer_ids = np.arange(1000, 1000 + n_customers) # genera un array de números consecutivos (enteros) usando NumPy
names = [f' customer_{i} ' for i in range(n_customers)]
cities = np.random.choice(['Madrid', 'Barcelona', 'Valencia', 'Sevilla'], size=n_customers)
segments = np.random.choice(['Standard', 'Premium'], size=n_customers, p=[0.75, 0.25])
ages = np.random.normal(39, 12, size=n_customers).round().astype(float)

df_customers = pd.DataFrame({
    'customer_id': customer_ids,
    'name':names,
    'city':cities,
    'segment':segments,
    'age':ages
})

# Ensuciar DataFrame Customers
# np.random.choice(df_customers.index, 10, replace=False) Selecciona 10 filas aleatorias, y replace=False evita repetir filas.

# Poner valores Nan en la columna age
df_customers.loc[np.random.choice(df_customers.index, 10, replace=False), 'age'] = np.nan

# Poner valores None en la columna city
df_customers.loc[np.random.choice(df_customers.index, 6, replace=False), 'city'] = None

# Concatenera filas al dataframe
# df_customers.sample(8, random_state=42) selecciona 8 filas aleatorias
# random_state=42 que siempre sean las mismas.
# ignore_index=True reinicia los índices
df_customers = pd.concat([df_customers, df_customers.sample(8, random_state=42)], ignore_index=True)

print("Customers.Shape: ", df_customers.shape)
display("DataFrame Customers", df_customers.head())
```

### DataFrame Products
Columnas: 'product_id', 'product_name', 'category', 'base_price'

```python
# Crear DataFrame Products
products_id = np.arange(200, 240)
products_name = [f' Product_{pid} ' for pid in products_id]
categories = np.random.choice(['Electronics', 'Home', 'Sports', 'Books'], size=len(products_id))
base_prices = np.random.uniform(5, 250, size=len(products_id)).round(2)


df_products = pd.DataFrame({
    'product_id':products_id,
    'product_name':products_name,
    'category':categories,
    'base_price':base_prices,
})

# Ensuciar DataFrame Products
df_products.loc[np.random.choice(df_products.index, 5, replace=False), 'category'] = ' electronics '
df_products.loc[np.random.choice(df_products.index, 4, replace=False), 'base_price'] = np.nan
df_products = pd.concat([df_products, df_products.sample(5, random_state=1)], ignore_index=True)

print("Products.Shape: ", df_products.shape)
display("DataFrame Products", df_products.head())
```

### DataFrame Transactions:
#### Transactions(Pedidos- un consumidor hace pedido de productos) ---
Columnas: 'tx_id', 'customer_id','product_id','quantity','discount','tx_date'

```python
# Crear DataFrame Transactions(Pedidos)
n_tx = 900
tx_ids = np.arange(5000, 5000 + n_tx)

tx_customer = np.random.choice(df_customers['customer_id'].unique(), size=n_tx)
tx_product = np.random.choice(df_products['product_id'].unique(), size=n_tx)
quantity = np.random.randint(1, 6, size=n_tx)
discount = np.random.choice([0.0, 0.05, 0.10, np.nan], size=n_tx, p=[0.55, 0.2, 0.15, 0.10])

dates = pd.to_datetime('2025-01-01') + pd.to_timedelta(np.random.randint(0, 90, size=n_tx), unit='D')

df_transactions = pd.DataFrame({
    'tx_id': tx_ids,
    'customer_id': tx_customer,
    'product_id': tx_product,
    'quantity': quantity,
    'discount': discount,
    'tx_date': dates,
})

# Ensuciar DataFrame Transaction
df_transactions.loc[np.random.choice(df_transactions.index, 7, replace=False), 'product_id'] = np.nan
transactions = pd.concat([df_transactions, df_transactions.sample(12, random_state=7)], ignore_index=True)

print("Transaction.Shape: ",df_transactions.shape)
display("DataFrame Transaction(Pedidos)", df_transactions.head())
```
