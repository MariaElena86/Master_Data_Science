### Ejercicio 5: Filtrar utilizando máscaras y crear columnas nuevas
mask(condiciones, pueden ser simples o con varias condiciones)

Usando `df_large`:
1. crea una columna `Total_AB` con `A + B`,
2. filtra filas con `A > 50`,
3. filtra filas con `A > 50` y `B < 20`,
4. muestra solo las columnas `A`, `B` y `Total_AB` del resultado filtrado.

```python
df_large['Total_AB'] = df_large['A'] + df_large['B']
display("Nueva Columna Total_AB", df_large.head)
mask_simple = df_large['A']>50
display("Columnas donde A>50", df_large[mask_simple].head(5))
mask_varias = (df_large['A']>50) & (df_large['B']<20)
display("Columnas donde A>50 y B<20", df_large[mask_varias].head)

display("Solo Columnas A, B y Total_AB", df_large[['A', 'B', 'Total_AB']])
```

### Ejercicio 6: Ordena resultados por columnas de forma ASC o DESC
ASC(por defecto, ordena de menor a mayor) o DESC(de mayor a menor)

1. Ordena `df_large` por la columna `C` de mayor a menor.
2. Ordena `df_large` por `B` descendente y `D` ascendente.
3. Muestra las 5 primeras filas de cada resultado.

```python
display("Ordenar por Columna C - DESC(de mayor a menor)", df_large.sort_values(by='C', ascending=False).head(5))
display("Ordenar por Columna B-DESC y D-ASC", df_large.sort_values(by=['B','D'], ascending=[False,True]).head(5))
```
