### 1. Diccionario de traducciones

Imagina que estás creando un pequeño **diccionario bilingüe** (Español → Inglés).  
1. Crea un diccionario llamado `traducciones` con **3 palabras** como claves en español (gato, perro, mesa) y sus traducciones en inglés (cat, dog, table) como valores.  
2. Muestra todas las **claves** (palabras en español) al usuario y pídele que introduzca una palabra en español (con `input()`).  
3. Comprueba si la palabra está en el diccionario:  
   - Si **existe**, muestra su traducción en inglés.  
   - Si **no existe**, indica que no se encuentra en el diccionario.  
4. Pregunta al usuario si desea **agregar** esa palabra al diccionario. Si responde que sí, pídele la traducción y añádela al diccionario.  
5. Muestra el diccionario final.

*Pistas:*
- Usa `in` para verificar si una clave existe en el diccionario.  
- Recuerda que para **insertar** o **actualizar** un par clave-valor se puede hacer con `diccionario[clave] = valor`.


````python
"""
Pasos:
- crear diccionario con valores iniciales
- mostrar todas las claves
- pedir al usuario que entre una palabra en español
- verificar si existe la palabra
  - existe: mostrar la traduccion
  - no existe: no se encuentra
    - pedir al usuario si quiere agregarla
      - si quiere: pedir que entre la traduccion y agregar al diccionario
- mostrar el diccionario final
 """

traducciones = {
    "gato":"cat",
    "perro":"dog",
    "mesa":"table"
  }
print("Las palabras en el diccionario son: ")
for clave in traducciones.keys():
    print(clave)


palabra = input("Escribe una palabra en español.").lower()
if palabra in traducciones.keys():
    print("La palabra existe, su traduccion es ", traducciones[palabra])
else:
    print(f"La palabra {palabra} no existe")
    deseo = input(f"Desea agregar la palabra '{palabra}'? Escribe si/no")
    if deseo == "si":
        traduccion_palabra = input(f"Escribe la traduccion de {palabra}")
        traducciones[palabra] = traduccion_palabra

print("Diccionario Final: ")
for clave, valor in traducciones.items():
    print(f"{clave}: {valor}")
````

### 2. Registro de notas de alumnos

Quieres llevar un pequeño **registro** de los alumnos y sus **notas**.  
1. Crea un diccionario vacío llamado `notas_alumnos`.  
2. En un **bucle** `while True`, pide al usuario que introduzca:  
   - El nombre del alumno.  
   - La nota del alumno (un número).  
3. Almacena esa información en el diccionario `notas_alumnos` (la clave será el nombre y el valor será la nota).  
4. Pregunta al usuario si desea **agregar otro alumno**.  
   - Si contesta `"no"` (o `"No"`, ignorando mayúsculas/minúsculas), rompe el bucle.  
5. Después de salir del bucle, recorre el diccionario y **muestra** por pantalla un mensaje del tipo:  




````python
"""
Pasos:
- crear diccionario vacio
- crear bucle while
  - pedir al usuario que entre nombre y nota
  - adicionar al diccionario {nombre:nota}
  - preguntar si desea agregar nuevo alumno
    - no/No: salir del bucle 
- mostrar mensaje(NO SE CUAL)
"""
notas_alumnos = {}
while True:
    nombre = input("Entra el nombre del alumno.")
    nota = input(f"Entra la nota para el alumno: {nombre}")
    notas_alumnos[nombre] = nota

    continuar = input("Desea agregar nuevo alumno? Escribe no/No:")
    if continuar in "no/No":
        break

print("Registro de notas: ", notas_alumnos)
````

### 3. Estadísticas de caracteres en una cadena

1. Pide al usuario que escriba una **frase** (o varias).  
2. Crea un diccionario llamado `conteo_caracteres` que guarde la **cantidad de veces** que aparece cada carácter en la frase.  
- Para cada carácter en la frase:  
  - Si ya existe como clave, incrementa su valor.  
  - Si no, añádelo con valor `1`.  
3. Ignora mayúsculas/minúsculas convirtiendo la frase a minúsculas antes de contar.  
4. Muestra el resultado en pantalla, por ejemplo:  
5. (Opcional) Excluye los **espacios** en blanco de la cuenta si lo deseas (puedes usar un condicional que no cuente los espacios).


````python
"""Pasos:
- pedir que entre una frase
- convertir a minuscula
- crear nuevo diccionario
- recorrer la frase y ver cuantas veces se encuentra cada caracter
  guardar en el diccionario {caracter:cantidad_veces}
  - si se repite aumentar el valor
- mostrar resultado 
"""
conteo_caracteres = {}
frase = input("Escriba una frase")
for caracter in frase.lower().replace(" ",""):
    if caracter in conteo_caracteres.keys():
            conteo_caracteres[caracter] += 1
    else:
        conteo_caracteres[caracter] = 1   
        
print(f"Frase: {frase}")
print(f"Diccionario:")
for clave, valor in conteo_caracteres.items():
    print(f"{clave}: {valor}")
````

### 4. Control de inventario

Imagina que gestionas un almacén con productos y cantidades en stock.  
1. Define un diccionario `inventario` con algunas **claves** como `"manzanas"`, `"naranjas"`, `"leche"`, asociadas a números que representen la **cantidad** disponible.  
2. Muestra un **menú** (usando un bucle `while`) con opciones:  
1. `"1. Ver inventario"`  
2. `"2. Añadir producto"`  
3. `"3. Actualizar stock de un producto"`  
4. `"4. Eliminar producto"`  
5. `"5. Salir"`  
3. Según la opción elegida:  
- **Ver inventario**: Recorre el diccionario e imprime producto y cantidad.  
- **Añadir producto**: Pide el nombre del producto y la cantidad. Añádelo al diccionario (si ya existía, avisa al usuario o actualiza la cantidad).  
- **Actualizar stock**: Pide el nombre del producto. Si existe, pide la nueva cantidad y actualiza el valor en el diccionario. Si no existe, indícalo por pantalla.  
- **Eliminar producto**: Pide el nombre del producto y, si existe en el diccionario, elimínalo. Si no existe, indícalo.  
- **Salir**: Rompe el bucle.  
4. Muestra el diccionario final cuando el usuario salga.

````python
inventario = {"manzana":5, "naranja": 8, "leche": 3}
print("Menú: " \
    "\n 1. Ver Inventario " \
    "\n 2. Añadir producto" \
    "\n 3. Actualizar stock de un producto" \
    "\n 4. Eliminar producto" \
    "\n 5. Salir")

while True:
    opcion = int(input("Selecciona una opcion de menú"))
    if opcion == 1:#Ver Inventario
        for producto, stock in inventario.items():
            print(f"Producto: {producto}, Stock: {stock}")

    elif opcion == 2:#Añadir producto
        nuevo_producto = input("Escribe el nombre del producto")
        cantidad = input("Escribe la cantidad del producto")
        if nuevo_producto in inventario.keys():            
            inventario[nuevo_producto] += cantidad
        else:
            inventario[nuevo_producto] = cantidad

    elif opcion == 3:#Actualizar stock de un producto
        producto_existente = input("Escribe el nombre del producto que desea actualizar")
        if producto_existente in inventario.keys():  
            nuevo_stock = input("Escribe la nueva cantidad")          
            inventario[producto_existente] = nuevo_stock
        else:
            print("El producto no existe")

    elif opcion == 4:#Eliminar producto
        producto_existente = input("Escribe el nombre del producto que desea eliminar")
        if producto_existente in inventario.keys():  
            inventario.pop(producto_existente)
        else:
            print("El producto no existe")       
    elif opcion == 5:#Salir
        print("")
        break
    print("Inventario ", inventario.items()) 
````

```python
precios = {
```

2. Añade un nuevo producto "yogur" con un precio de 1.2.

3. Cambia el precio de las "manzanas" a 1.75.

4. Elimina el producto "leche" de la lista porque ya no lo vas a vender.

Muestra todos los productos disponibles (claves) y sus precios (valores) de manera clara.
