### Ejercicio 1: Utilizar (`map`)

- Calcula el promedio de notas por estudiante y genera una lista de tuplas con formato:
`(name, average)`

```python
# Ejercicio 1: Calcular promedio 

students = [
    {"name": "Ana", "grades": [8, 7, 9], "absences": 1},
    {"name": "Luis", "grades": [5, 6, 4], "absences": 3},
    {"name": "Marta", "grades": [9, 9, 10], "absences": 0},
    {"name": "Juan", "grades": [6, 7, 6], "absences": 2},
    {"name": "Sofia", "grades": [4, 5, 6], "absences": 4},
]

def average(grades):
    return sum(grades) / len(grades)

#Version usando bucle
list_averages = []
for a in students:
    list_averages.append((a["name"], round(average(a["grades"]),2)))
print(list_averages)

#Version usando lambda y map()
map_average = list(map(
    lambda student: (
        (student["name"], round(average(student["grades"]), 2))
    ),
    students 
))
print(map_average)
```

## Ejercicio 2: Utilizar (`filter`): 
### Filtrar los mejores estudiantes

- Filtra los estudiantes que cumplan las dos condiciones: 
  - promedio `>= 6`
  - ausencias `<= 2`

```python
# Exercise 2 -- Filtrar los mejores estudiantes
estudiantes = [
    {"name": "Ana", "grades": [8, 7, 9], "absences": 1},
    {"name": "Luis", "grades": [5, 6, 4], "absences": 3},
    {"name": "Marta", "grades": [9, 9, 10], "absences": 0},
    {"name": "Juan", "grades": [6, 7, 6], "absences": 2},
    {"name": "Sofia", "grades": [4, 5, 6], "absences": 4},
]

#VERSION A -- Usando for y append
# Recorrer la lista de estudiantes e ir guardando en una nueva el que cumpla las dos condiciones
best_students = []

for student in estudiantes:
    grades = student["grades"]
    absences = student["absences"]
    average = round(sum(grades) / len(grades), 2)
    if average >= 6 and absences <= 2:
        best_students.append(({"estudiante":student["name"]}, 
                               {"Average":average}, 
                               {"Ausencias": student["absences"]}
                               ))

print("VERSION A -- Mejores estudiantes(aprobados, sin ausencias):")
for s in best_students:
    print(s)

#VERSION B -- Usando filter dos veces
# Paso 1: filtrar la lista por los aprobados
def average(estudiante):
    grades = estudiante["grades"]
    return round(sum(grades) / len(grades), 2)

aprobados = list(filter(average, estudiantes))
print("Estudiantes aprobados:")
for s in aprobados:
    print(f"Estudiante: {s["name"]} -- Notas: {s["grades"]}")

# Paso 2: filtrar la lista de aprobados filtro por ausencias
mejores_estudiantes = list(filter(lambda estudiante: estudiante["absences"] <= 2, aprobados))
print("VERSION B -- Mejores estudiantes(aprobados, sin ausencias):")
for s in mejores_estudiantes:
    print(f"Estudiante: {s["name"]} -- Notas: {s["grades"]} -- Ausencias: {s["absences"]}")


#VERSION C -- Usando filter una sola vez
def find_best_student(student):
    grades = student["grades"]
    absences = student["absences"]
    average = round(sum(grades) / len(grades), 2)
    return average >= 6 and absences <= 2

best_students = list(filter(find_best_student, estudiantes))
print("VERSION C -- Mejores estudiantes(aprobados, sin ausencias):")
for s in best_students:
    print(f"Estudiante: {s["name"]} -- Notas: {s["grades"]} -- Ausencias: {s["absences"]}")
```

### Ejercicio 4 : Utilizar (`sorted`)
Genera una lista con los 3 mejores estudiantes usando el mismo criterio de ordenación:
- mejor promedio primero,
- en caso de empate, menos ausencias.

```python
# Ejercicio 4 - Utilizar sorted(..., key=..)
students = [
    {"name": "Ana", "grades": [8, 7, 9], "absences": 1},
    {"name": "Luis", "grades": [5, 6, 4], "absences": 3},
    {"name": "Marta", "grades": [9, 9, 10], "absences": 0},
    {"name": "Juan", "grades": [6, 7, 6], "absences": 2},
    {"name": "Sofia", "grades": [4, 5, 6], "absences": 4},
]
def average(grades):
    return sum(grades) / len(grades)

ranked_students = sorted(
    students,
    key=lambda student: (-average(student["grades"]), student["absences"]),
)[:3]
for student in ranked_students:
    print(f"{student['name']}: Average = {average(student['grades']):.2f}, Absences = {student['absences']}")
```

# Ejercicio 5: Mini-pipeline

A partir de un dataset con entradas inválidas:

1. Elimina estudiantes sin notas válidas.
2. Calcula el promedio y crea un nuevo diccionario con `name`, `average` y `absences`.
3. Filtra estudiantes con promedio `>= 6` y ausencias `<= 2`.
4. Ordena por promedio descendente y ausencias ascendente.


```python
# Ejercicio 5 -- Mini-pipeline
raw_students = [
    {"name": "Ana", "grades": [8, None, 9], "absences": 1},
    {"name": "Luis", "grades": [], "absences": 3},
    {"name": "Marta", "grades": [9, 9, -10], "absences": 0},
    {"name": "Juan", "grades": None, "absences": 2},
    {"name": "Sofia", "grades": [4, 5, 6], "absences": 4},
    {"name": "Elena", "grades": [10, 9, 9], "absences": 0},
]
def average(grades):
    valid_grades = list(filter(lambda x: x is not None and x > 0 ,grades))
    if len(valid_grades) > 0:    
        return sum(valid_grades) / len(valid_grades)

# 1- Eliminar alumnos con notas no validas:
#  - el array grades no es None
#  - las notas dentro de grades son >0
def notas_validas(student):
    grades = student["grades"]
    if grades is not None and len(grades) > 0:
        lmd = lambda x: x is not None and x > 0 # buscar notas validas en grades
        validgrades = list(filter(lmd, grades))
        if len(validgrades) == len(grades): # Todas las notas en grades deben ser válidas
            return True
    return False

valid_students = list(filter(notas_validas, raw_students))
print("Estudiante con notas validas:")
for s in valid_students:
    print(s)

# 2- Calcula el promedio y crea un nuevo diccionario con `name`, `average` y `absences`.
fn = lambda student: {
            "name": student["name"],
            "average": average(student["grades"]),
            "absences": student["absences"],
        }
with_average = list(map(fn,valid_students,))
print("Nuevo diccionario con columna average:")
for s in with_average:
    print(s)

#3 - Filtra valid_students que tengan promedio `>= 6` y ausencias `<= 2`.
fn = lambda student: (average(student["grades"]) >= 6 and student["absences"] <= 2)
eligible_students = list(filter(fn, valid_students))

print("Estudiantes con promedio `>= 6` y ausencias `<= 2:")
for s in eligible_students:
    print(s)
    
#4 - Ordena por promedio descendente y ausencias ascendente.
students = [
    {"name": "Ana", "grades": [8, 6, 9], "absences": 1},
    {"name": "Luis", "grades": [9, 9, 10], "absences": 3},
    {"name": "Marta", "grades": [9, 8, 10], "absences": 0},
    {"name": "Juan", "grades": [8, 5, 6], "absences": 2},
    {"name": "Sofia", "grades": [4, 5, 6], "absences": 4},
    {"name": "Elena", "grades": [10, 9, 9], "absences": 0},
]
orderfn = lambda s: (-average(s["grades"]), s["absences"])
ranked_students = sorted(students, key=orderfn)

print("Order By average-DESC y absences-ASC")
for s in eligible_students:
    print(s)
```

# Ejercicio 6: Reto opcional: lista de honor

Crea un pipeline que devuelva solo los nombres de estudiantes que cumplan:
- promedio `>= 8.5`
- ausencias `== 0`
Intenta hacerlo con pasos pequeños y legibles.

```python
# Ejercicio 6 -- Reto opcional
students = [
    {"name": "Ana", "grades": [8, 6, 9], "absences": 1},
    {"name": "Luis", "grades": [9, 9, 10], "absences": 3},
    {"name": "Marta", "grades": [9, 8, 10], "absences": 0},
    {"name": "Juan", "grades": [8, 5, 6], "absences": 2},
    {"name": "Sofia", "grades": [4, 5, 6], "absences": 4},
    {"name": "Elena", "grades": [10, 9, 9], "absences": 0},
]
def average(grades):
    return sum(grades) / len(grades)

def find_best_student(student):
    grades = student["grades"]
    absences = student["absences"]
    average = round(sum(grades) / len(grades), 2)
    return average >= 8.5 and absences == 0

best_students = list(filter(find_best_student, students))

print("Estudiantes con  average >= 8.5 y ausencias==0")
for s in best_students:
    print({
            "name": s["name"],
            "average": round(average(s["grades"]),2),
            "absences": s["absences"],
        })
```
