### 1️⃣ Calcular el total de ventas del día

#### Objetivo:
Un comercio registra sus ventas una a una. 
El programa debe permitir al usuario ingresar el valor de cada venta.  
El proceso termina cuando se introduce un valor de venta igual a `0`.
Al final, se muestra el total acumulado.

```python
total = 0

while True:
    venta = float(input("Venta (0 para terminar): "))
    
    if venta == 0:
        break
    
    total += venta

print("Total:", total)
```

### 2️⃣ Registrar pagos hasta alcanzar una meta

#### Objetivo:
Una tienda necesita recaudar al menos **500 euros** en el día.  
Cada cliente paga una cantidad y se va sumando al total.  
El programa debe seguir pidiendo montos hasta que se alcance (o supere) la meta.  
Al final, muestra cuánto se recaudó y cuántos pagos fueron necesarios.

```python
total_recaudado = 0
total_pagos = 0
meta = 500

while total_recaudado < meta:
    pago = float(input('Inserta el pago'))
    total_recaudado += pago
    total_pagos += 1

print(f'Ha recaudado un total de {total_recaudado} en {total_pagos} pagos')
```

### 3️⃣ Control de stock de productos
##### Objetivo:
Un sistema de inventario debe restar unidades del stock cada vez que hay una venta.
El stock inicial es 20. En cada iteración, se ingresa cuántas unidades se vendieron.
El bucle se detiene cuando no queda stock suficiente para cubrir la venta.

```python
stock = 20
hay_stock = True

while stock > 0 and hay_stock:
    venta = int(input('Ingrese el total a vender'))

    if venta > stock:
        print(f'No hay suficiente stock. intenta vender {venta} y solo tiene en el stock {stock}')
        hay_stock=False
    else:    
        stock -= venta
        print("Venta realizada.")
    
print('Fin de las ventas')
```

### 4️⃣ Registrar empleados hasta llegar al límite
##### Objetivo:
Una empresa va registrando empleados nuevos. El sistema pide el nombre de cada uno.
Solo se pueden registrar hasta 10 empleados. Cuando se llega a ese número, se detiene el registro y se imprime cuántos se registraron en total.

```python
lista_empleados = []
registrados = 0
limite = 10

while registrados < limite:
    empleado = input('Inserte el nombre del empleado')
    lista_empleados.append(empleado)
    registrados += 1

print('Total registrados: ', registrados)
print('Empleados: ', lista_empleados)

```
