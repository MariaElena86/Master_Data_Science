## 1. Sintaxis bĆ”sica: `try-except`

Es el mecanismo principal para evitar que un programa se detenga abruptamente.

- `try`: aquĆ­ va el cĆ³digo que puede fallar.
- `except`: aquĆ­ decides quĆ© hacer si ocurre un error.

### Errores comunes
- `ValueError`: el tipo parece correcto, pero el contenido no lo es.
- `ZeroDivisionError`: dividir por cero.
- `TypeError`: mezclar tipos incompatibles.
- `IndexError`: acceder a una posiciĆ³n que no existe.


```python
# Ejemplo 2: capturar errores frecuentes
try:
    number1 = int(input("Enter a number as a base: "))
    number2 = int(input("Enter another number as a divisor: "))
    result = number1 / number2
    print(f"The result of {number1} divided by {number2} is {result}")
except ValueError:
    print("Error: Invalid input. Please enter numeric values.")
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
# Ejemplo 3: otros errores tĆ­picos
try:
    text = "Hello, World! "
    number = 5
    combined = text + number
    print(combined)
except TypeError:
    print("Error: Cannot concatenate string and integer.")

try:
    my_list = [1, 2, 3]
    print(my_list[5])
except IndexError:
    print("Error: Index is out of range.")
```

## 2. El ciclo completo: `else` y `finally`

Podemos extender la estructura para tener mĆ”s control:

- `else`: se ejecuta solo si **no** hubo error.
- `finally`: se ejecuta siempre, haya error o no.

`finally` es Ćŗtil para limpieza; `else` ayuda a separar el caso exitoso del caso con error.


```python
# Ejemplo 4: else y finally
try:
    result = 10 / 5
except ZeroDivisionError:
    print("Cannot divide by zero.")
else:
    print(f"Operation successful. Result: {result}")
finally:
    print("Block finished.")

print("Continuing with the rest of the program.")
```

## 3. ValidaciĆ³n con `raise`

Usamos `raise` cuando el cĆ³digo es vĆ”lido para Python, pero invĆ”lido para nuestra lĆ³gica.

Ejemplo tĆ­pico: edad negativa, precio negativo o mes fuera de rango.


```python
def set_age(age: int):
    if age <= 0:
        raise ValueError("Age must be a positive integer.")
    print(f"Age set to {age}")

try:
    set_age(int(input("Enter your age: ")))
except ValueError as e:
    print(f"Validation Error: {e}")
```

## 4. PatrĆ³n muy Ćŗtil: input robusto con `while True`

Cuando pides datos al usuario, no quieres que el programa muera a la primera entrada incorrecta.

Este patrĆ³n es muy comĆŗn en scripts y pequeĆ±os menĆŗs por terminal.


```python
def get_valid_int():
    while True:
        try:
            return int(input("Enter a number: "))
        except ValueError:
            print("Invalid input. Try again.")

number = get_valid_int()
print(f"You entered: {number}")
```

## Ejercicio 2: Conversor de Lista

Convierte esta lista a enteros.

Objetivo:
- ignorar los valores que den error
- seguir procesando el resto
- guardar los vĆ”lidos en una nueva lista


```python
# Your code here
lista = ["10", "20", "hola", "30", "4.5", None, "50"]

nueva_lista = []

for elemento in lista:
    try:
        if int(elemento):
            nueva_lista.append(elemento)
    except (ValueError, TypeError):
        # Ignora el error y continĆŗa
        continue

print(f"Lista fina: {nueva_lista}")


# Version con comprehension
lista2 = ["10", "20", "hola", "30", "4.5", None, "50"]

nueva_lista2 = [ 
    int(x)
    for x in lista2
    if isinstance(x, (int, str)) and str(x).isdigit()
]

print(f"Lista fina2: {nueva_lista2}")

```
