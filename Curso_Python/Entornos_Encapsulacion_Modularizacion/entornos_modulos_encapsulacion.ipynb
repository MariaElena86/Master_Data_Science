## 1. Entornos y Scope (La regla LEGB)

Python busca las variables en un orden específico. Si no entiende esto, tendrás errores de "Variable not defined" o sobrescribirás datos importantes.

El orden es **LEGB**:
1.  **L**ocal: Dentro de la función actual.
2.  **E**nclosing: En la función padre (si hay funciones anidadas).
3.  **G**lobal: En el nivel superior del archivo.
4.  **B**uilt-in: Funciones internas de Python (como `print`, `len`).

Lo importante: **Las variables dentro de una función mueren cuando la función termina.**

```python
# Ejemplo de Scope
vidas = 3  # Variable GLOBAL

def jugar():
    enemigos = 5  # Variable LOCAL
    # Podemos LEER la global, pero no modificarla directamente
    print(f"Enemigos: {enemigos}, Vidas: {vidas}")

jugar()
# print(enemigos) # Esto daría error porque 'enemigos' ya no existe
```

## 3. Métodos Estáticos (`@staticmethod`)

**Teoría:** A veces queremos meter una función dentro de una clase para mantener el orden, pero esa función **no necesita usar `self`** (no toca los datos del objeto). Es como una herramienta útil.

Usamos el decorador `@staticmethod`.

```python
class Calculadora:
    @staticmethod
    def sumar(a, b):
        return a + b

# Fíjate: ¡No hace falta crear un objeto con Calculadora()!
print(Calculadora.sumar(5, 10))
```

## 4. Encapsulación Básica: Niveles de Privacidad

En Python, a diferencia de Java o C++, la privacidad se basa en **convenciones** y trucos de nombres.

1.  **Público (`nombre`)**: Cualquiera puede acceder y modificar. Es lo normal.
2.  **Protegido (`_nombre`)**: Empieza por un guion bajo. Significa: *"Cuidado, esto es interno, no lo toques si no sabes qué haces"*. (Pero Python te deja tocarlo).
3.  **Privado (`__nombre`)**: Empieza por **dos** guiones bajos. Python cambia el nombre internamente (Name Mangling) para que sea muy difícil acceder desde fuera.

```python
class SecretFile:
    def __init__(self, filename, content):
        self.filename = filename       # Public: Everyone can see the name
        self._backup = True            # Protected: Internal flag
        self.__content = content       # Private: The sensitive info

file = SecretFile("plans.txt", "Attack at dawn")

print("Filename:", file.filename)    # ✅ OK
print("Backup:", file._backup)     # ⚠️ Works, but frowned upon
#print("Content:", file.__content) # ❌ AttributeError: 'SecretFile' object has no attribute '__content'
```

## 5. Controlando el Acceso: `@property`

Ahora tenemos un problema: si hacemos una variable privada (`__saldo`), **nadie** puede leerla ni modificarla desde fuera. ¿Y si queremos dejar que la lean pero no que la cambien? ¿O validar antes de cambiar?

Usamos **Getters** y **Setters** con el decorador `@property`.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self._age = age  # We protect it

    # GETTER: Allows reading 'p.age' as a variable
    @property
    def age(self):
        return self._age

    # SETTER: Allows assigning 'p.age = 50' with validation
    @age.setter
    def age(self, new_age):
        if new_age < 0:
            print("❌ Age cannot be negative!")
        else:
            print("✅ Age updated")
            self._age = new_age

p = Person("Alice", 30)
print(p.age)   # Calls the Getter
p.age = -10    # Calls the Setter (Validation error)
p.age = 35     # Calls the Setter (Success)
```

### ⚡ Ejercicio Rápido 5: Login Seguro
Ahora sí, vamos a proteger un usuario.
1. Crea una clase `User` con `username` y `__password` (privado).
2. Crea un **Getter** para `password` que devuelva siempre `"*******"` (ocultando la real).
3. Crea un **Setter** para `password` que solo permita cambiarla si la nueva tiene más de 6 caracteres.
4. Prueba a leer la contraseña y a cambiarla.

```python
# Tu clase User aquí
class Usuario:

    def __init__(self, username, password):
        self.username = username
        self.__password = password #private
    
    @property
    def password(self): #Getter: permite leer un atributo que es privado
        return "*" * len(self.__password)
    
    @password.setter #Setter: permite modificar un atributo que es privado
    def password(self, new_password):
        if len(new_password) <= 6:
            print("❌ La contraseña debe tener al menos 7 caracteres.")
        else:   
            self.__password = new_password
            print("✅ Contraseña actualizada.")

user = Usuario("Juan", "1234")

print("Nombre de ususario: ",user.username)
print("Contraseña: ",user.password)
user.password = "22334455"
print("Nueva Contraseña: ",user.password)
```

## 🏆 PROYECTO FINAL: Sistema Modular de Empleados

Vamos a unirlo todo: Modularización, Encapsulación y Métodos Estáticos.

**Paso 1:** Ejecuta la celda de abajo para crear el archivo `empleados.py`. Analiza el código: tiene atributos privados, setters con validación y un método estático.

```python
%%writefile funciones/empleados.py

class Employee:
    def __init__(self, name, salary):
        self._name = name #protected
        self.__salary = salary#private

    @property#Getter, para leer el atributo protected
    def name(self):
        return self._name

    @property#Getter, para leer el atributo private
    def salary(self):
        return self.__salary

    @salary.setter#para modificar el atributo privado
    def salary(self, amount):
        if amount < 0:
            print("❌ Error: El salario no puede ser negativo.")
        else:
            print(f"✅ Salario actualizado a {amount}")
            self.__salary = amount

    @staticmethod
    def calculate_tax(amount):
        return amount * 0.15
```
