```python
#Convertir el agente en uno predictivo con scikit-learn
#Ahora el modelo aprenderá “con estos patrones → probablemente conviene promocionar o no”
import pandas as pd

df = pd.read_csv("ventas.csv")
print(df)

# Creamos lo que el modelo debe aprender a predecir.
df["decision"] = df["total_sale"].apply(lambda x: 1 if x > 100 else 0)
# Donde 1 = “buena venta” y 0 = “mala venta”

# Preparar datos para entrenar un modelo de IA.
# X e y (lo más importante de todo ML) donde 
# X = variables de entrada (features, Son los datos que el modelo usa para aprender) y
# y = lo que quieres predecir(Es la “respuesta correcta”)

from sklearn.model_selection import train_test_split 
#es una funcion de la libreria scikit'learn, su trabajo es dividir datos en entrenamiento y prueba
#train_test_split divide los datos en dos partes
# 1era parte: TRAIN (entrenamiento)
# 80% de los datos, el modelo aprende aquí
# 2da parte: TEST (prueba)
# 20% de los datos, se usa para evaluar el modelo
#Ejemplo Si tienes 100 filas: 
# 80 filas → se usan para entrenamiento y 20 → prueba


#Preparar datos para IA
X = df[["price", "stock", "total_sale"]]
y = df["decision"]

X_train, X_test, y_train, y_test = train_test_split(
    X, #información
    y, #resultado que quieres predecir
    test_size=0.2, #significa: 20% para test, 80% para entrenamiento
    random_state=42 # siempre divide los datos igual
)
# Salida de la funcion train_test_split
# X_train  # datos de entrada para entrenar
# X_test   # datos de entrada para probar
# y_train  # resultados correctos para entrenar
# y_test   # resultados correctos para evaluar

#Crear el modelo
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

#RandomForestClassifier es un modelo de scikit-learn,
# usa muchos árboles de decisión, 
# combina sus resultados, 
# toma una decisión final más fiable

#Crear el modelo nuevo
modelo = RandomForestClassifier()

#Entrenar el modelo
modelo.fit(X_train, y_train) #decirle al modelo que aprenda

#Evaluar el modelo, te dice qué tan bueno es el agente
predicciones = modelo.predict(X_test)#para ver qué decisiones toma

#accuracy_score sirve para medir el rendimiento del modelo
#Mide: “¿Cuántas predicciones fueron correctas?”
#formula accuracy = aciertos / total
print("Accuracy:", accuracy_score(y_test, predicciones))
# Ejemplo de resultado Accuracy: 0.85 esto significa que el modelo acierta el 85% de las veces 
# es como un examen donde:
# predicciones → son tus respuestas 
# y_test → son las respuestas correctas
# accuracy → es la nota del examen


#Usar el agente predictivo
df["prediccion"] = modelo.predict(X)

print(df)




```
