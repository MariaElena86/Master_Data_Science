```python
#Ejercicio Practico
import pandas as pd

estudiantes = [
    {"nombre": "Ana Garcia", "matematicas": 8.5, "lengua": 7.2, "historia": 9.0, "ciencias": 8.8},
    {"nombre": "Luis Martinez", "matematicas": 6.0, "lengua": 6.5, "historia": 7.1, "ciencias": 6.8},
    {"nombre": "Marta Lopez", "matematicas": 9.2, "lengua": 8.7, "historia": 8.9, "ciencias": 9.5},
    {"nombre": "Carlos Sanchez", "matematicas": 5.8, "lengua": 6.1, "historia": 6.0, "ciencias": 5.5},
    {"nombre": "Elena Fernandez", "matematicas": 7.5, "lengua": 8.0, "historia": 7.8, "ciencias": 8.2},
    {"nombre": "Javier Perez", "matematicas": 4.9, "lengua": 5.5, "historia": 6.2, "ciencias": 5.8},
    {"nombre": "Laura Gomez", "matematicas": 8.8, "lengua": 9.1, "historia": 8.5, "ciencias": 9.0},
    {"nombre": "Diego Ruiz", "matematicas": 6.7, "lengua": 7.0, "historia": 6.5, "ciencias": 6.9},
    {"nombre": "Sofia Torres", "matematicas": 9.5, "lengua": 9.3, "historia": 9.7, "ciencias": 9.8},
    {"nombre": "Pablo Navarro", "matematicas": 7.2, "lengua": 6.8, "historia": 7.0, "ciencias": 7.5}
]

# Crear un DataFrame a partir de la lista de estudiantes
students_df = pd.DataFrame(estudiantes)

# 1- Calcular el promedio de cada estudiante por asignaturas
students_df['average'] = students_df[['matematicas', 'lengua', 'historia', 'ciencias']].mean(axis=1)

# 2- Filtrar los estudiantes con promedio mayor a 9
filtered_students_df = students_df[students_df['average'] > 9.0]
print(filtered_students_df.head()) 

# 3- Exportar el DataFrame filtrado a un archivo CSV
filtered_students_df.to_csv('filtered_students.csv', index=False)
```

```python
students_ds = [
    {"id": 1, "name": "Ana"},
    {"id": 2, "name": "Luis"}
]
grades_ds = [
    {"id": 2, "grade": 8.5},
    {"id": 3, "grade": 6.0}
]
students_df = pd.DataFrame(students_ds)
grades_df = pd.DataFrame(grades_ds)

#Merge dos DataFrame por la columna id con Inner Join - solo los que coinciden (exmplo el id 2)
merge_inner_df = pd.merge(students_df, grades_df, on='id', how='inner')
print(merge_inner_df)

#Merge con Outer Join
merge_outer_df = pd.merge(students_df, grades_df, on='id', how='outer')
print(merge_outer_df)
```

```python
#Ejercicio practico
#Crear dataset con datos sobre empleados(nombre, departamento y salario)
employes_dataset = {
    "name": ["Ana Lopez", "Luis Perez", "Carlos Ruiz", "Elena Gomez", "Marta Sanchez"],
    "departament": ["RRHH", "IT", "Finanzas", "Marketing", "IT"],
    "salary": [2500, 3200, 2800, 2700, 3500]
}

employs_df = pd.DataFrame(employes_dataset)

#1- Calcular el salario promedio por departamento
average_df = employs_df.groupby('departament')['salary'].mean()# Calcular el average por departamento
average_df.head()

#2 Filtrar los trabajadores que su salario es mayor que la media de su departamento
#Version 1
employs_df["average"] = employs_df["departament"].map(average_df)#adicionar una columna al df que tenga el average de cada dept
employs_df.head()

filtered_df = employs_df[employs_df["salary"] > employs_df["average"]]
filtered_df.head()

#Version2
filtered_df = employs_df[
    employs_df.apply(lambda row: row["salary"] > average_df[row["departament"]], axis=1)
]
filtered_df.head()

#3 Guardar el resultado en un file csv
filtered_df.to_csv('high_salary_employers', index=False)
```
