```python
def is_leap_year(year):
    try:
        año = int(year)
        if (año % 400 == 0 or (año % 4 == 0 and año % 100 != 0)):
            return True
        else:
            return False
    except Exception as e:
        print(f"Error: {e}")
        return None


is_leap_year(2010)  # True
```

```python
"""
Simulador de lanzamiento de dados
1- Utilizar la libreria random
2- Pedir al usuario cuantas veces quiere lanzar el dado
3- Usar la funcion roll_dice() para generar un numero aleatorio entre 1 y 6,
 simulando el lanzamiento de un dado de 6 caras
4- Imprimir el resultado de cada lanzamiento
"""
import random

def roll_dice():
    return random.randint(1, 6)

def lanzar_dados(num_lanzamientos):
    for _ in range(num_lanzamientos):
        numero_del_dado = roll_dice()
        print(f"El numero que muestra el dado es {numero_del_dado}")

intentos = int(input("Entre la cantidad de lanzamientos que desea realizar: "))
lanzar_dados(intentos)
```
